# Pandas Series
Series is one dimentional array with axis labels

```python
import numpy as np
import pandas as pd
```

```python
lablels = ['a', 'b', 'c']
my_list = [10, 20, 30]
arr = np.array([10, 20, 30])
d = {'a': 10, 'b': 20, 'c': 100}
```

```python
pd.Series(my_list)
```

```python
# inedx the series with the lables list
pd.Series(data=my_list, index=lablels)
```

```python
pd.Series(data=arr)
```

```python
pd.Series(arr, lablels)
```

```python
pd.Series(d)
```

```python
pd.Series(data=lablels)
```

```python
# parsing in functions into series
pd.Series([sum, print, len])
```

```python
ser1 = pd.Series([1, 2, 3, 4], index=['USA', 'CHINA','FRANCE','GERMANY'])
```

```python
ser1
```

```python
ser2 = pd.Series([1, 2, 3, 4], index=['USA', 'CHINA','ITALY','JAPAN'])
```

```python
ser2
```

```python
ser1['USA']
```

```python
ser2['JAPAN']
```

```python
# adding up two series - adds the values with the same index labels
ser1 + ser2
```
