A collection of series with a shared index

```python
import numpy as np
import pandas as pd
```

```python
from numpy.random import randn
```

```python
np.random.seed(101) # seed for reproducibility
```

pd.DataFrame(data, index, columns)

```python
df = pd.DataFrame(randn(5,4), ['A','B','C','D','E'], ['W','X','Y','Z'])
```

```python
df
```

```python
df['W']  # accessing a single column, returns a Series
```

```python
type(df['W'])
```

```python
type(df)
```

```python
df[['W','Z']]  # accessing multiple columns, returns a DataFrame    
```

```python
df['new'] = df['W'] + df['Y']  # creating a new column
```

```python
df
```

```python
df.drop('new', axis=1)  # dropping a column. axis=0 refers to rows and axis=1 refers to columns
```

```python
df
```

```python
df.drop('new', axis=1, inplace=True)  # dropping a column permanently. Prevents deletign by mistake
```

```python
df
```

```python
df.drop('E') # default axis is 0, so we can drop rows this way
```

```python
df.shape # is a 2d matrix
# so rows refer to axis 0 and columns refer to axis 1 
# (5, 4) - 5 rows and 4 columns
```

```python
# selecting rows
df.loc['A']  # accessing a single row
# note that both rows and columns are series
```

```python
df.iloc[2]  # access via integer location even when the index is alphanumeric
```

```python
# selecting subsets of rows and columns
df.loc['B','Y']  # accessing a specific value
```

```python
# passing in lists of row and column names
df.loc[['A','B'], ['W','Y']]
```

# Conditional selection

```python
# check for values in dataframe greater than 0
df > 0
```

```python
bulldf = df>0 
```

```python
bulldf
```

```python
# get the values that are greater than 0
df[bulldf]  # this will show NaN for values that are not greater than 0
```

```python
# simple way to get values greater than 0
df[df>0]
```

```python
df['W'] > 0  # boolean series showing True/False for values greater than 0 in column W
```

```python
df['W']
```

```python
# return the rows where column W's value is greater than 0
df[df['W']>0]
```

```python
# all the rows where Z is less than 0
df[df['Z']<0]
```

```python
resultdf = df[df['W']>0]    
```

```python
resultdf['X']
```

```python
df[df['W']>0]  # get all the rows where W is greater than 0
```

```python
# return the X column values where W is greater than 0
df[df['W']>0]['X']
```

```python
df[df['W']>0][['Y','X']]
```

```python
boolser = df['W']>0
result = df[boolser]
mycol = ['X','Y']
result = result[mycol]
```

```python
result
```

# Explore 2 or more conditions

```python
# use & and | with parentheses to explore multiple conditions
# Explore 2 or more conditions
df[(df['W']>0) & (df['Y']>1)]
```

```python
# use or condition
df[(df['W']>0) | (df['Y']>1)]
```
