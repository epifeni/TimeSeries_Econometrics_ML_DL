# Indexing a 2 dimentioanl array (matrix)

Indexing exmples <br/>
mat[row, col]<br/>
mat[row][col]<br/>

```python
mat = np.array([[5,10,15],
                [20,25,30],
                [35,40,45]])
mat
```

```python
mat[0]
```

```python
mat[2]
```

```python
mat[1][1]
```

```python
mat[1, 1]
```

## 2D Array Slicing

```python
# pick the row then the column - upto row 2 and all columns after column 1
mat[:2, 1:]
```

## Parsing a boolian array into an arry

```python
bull_arr = arr > 4
arr[bull_arr]
```

```python
# Commonly done in one line using conditional selection
arr[arr > 4]
```

```python
arr[arr <= 9]
```
